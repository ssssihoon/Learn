## 1.고객별 분석

```python
# 전체 고객수
customer_df['고객ID'].unique()


# row데이터가 5000 있으므로 전체 고객수는 5000.
```

```python
# 월별 활동고객수
store_df['거래일자'] = pd.to_datetime(store_df['거래일자'], format='%Y%m%d')
store_df['월'] = store_df['거래일자'].dt.month
monthly_active_customers = store_df.groupby('월')['고객ID'].nunique().reset_index()
monthly_active_customers.columns = ['월', '활동고객수']

print(monthly_active_customers)
```

```python
# 성별 / 연령별 분포
gender_distribution = customer_df['성별'].value_counts()
print(gender_distribution)
print('=================\n')
age_distribution = (customer_df['연령'].value_counts().sort_values())
print(age_distribution)
```

```python
# 지역별 고객분포
area_list = customer_df['거주지역'].str.split()
customer_df_city_only = pd.DataFrame({'시': area_list.str[0]})
print(customer_df_city_only)
customer_df['시'] = area_list.str[0]
customer_df['시'].value_counts()
```

```python
# 일별 / 월별 구매고객수 추이

daily_customer_count = sales_df.groupby(['월', '일']).size()
print("일별 구매고객수:")
print(daily_customer_count)

monthly_customer_count = sales_df.groupby('월').size()
print("\n월별 구매고객수:")
print(monthly_customer_count)
```

```python
# gpt를 사용 << 일별 / 월별 구매고객수 추이

# 거래일자를 날짜형으로 변환
sales_df['거래일자'] = pd.to_datetime(sales_df['거래일자'], format='%Y%m%d')

# 요일 정보 추가 (0: 월요일, 1: 화요일, ..., 6: 일요일)
sales_df['요일'] = sales_df['거래일자'].dt.dayofweek

# 요일을 평일(0~4)과 주말(5,6)로 구분
sales_df['주중여부'] = sales_df['요일'].apply(lambda x: '평일' if x < 5 else '주말')

# 월 정보 추가
sales_df['월'] = sales_df['거래일자'].dt.month

# 계절 정보 추가
def get_season(month):
    if 3 <= month <= 5:
        return '봄'
    elif 6 <= month <= 8:
        return '여름'
    elif 9 <= month <= 11:
        return '가을'
    else:
        return '겨울'

sales_df['계절'] = sales_df['월'].apply(get_season)

# 일별, 월별 구매고객수 계산
customer_count_by_day = sales_df.groupby('거래일자').size()
customer_count_by_month = sales_df.groupby('월').size()

print("일별 구매고객수:")
print(customer_count_by_day)

print("\n월별 구매고객수:")
print(customer_count_by_month)

# 주중/주말과 계절에 따른 구매고객수 계산
customer_count_by_weekday = sales_df.groupby('주중여부').size()
customer_count_by_season = sales_df.groupby('계절').size()

print("\n주중/주말에 따른 구매고객수:")
print(customer_count_by_weekday)

print("\n계절에 따른 구매고객수:")
print(customer_count_by_season)
```

```python
# gpt를 사용 << 일별 / 월별 구매금액 추이
# 거래일자를 날짜형으로 변환
sales_df['거래일자'] = pd.to_datetime(sales_df['거래일자'], format='%Y%m%d')

# 요일 정보 추가 (0: 월요일, 1: 화요일, ..., 6: 일요일)
sales_df['요일'] = sales_df['거래일자'].dt.dayofweek

# 요일을 평일(0~4)과 주말(5,6)로 구분
sales_df['주중여부'] = sales_df['요일'].apply(lambda x: '평일' if x < 5 else '주말')

# 월 정보 추가
sales_df['월'] = sales_df['거래일자'].dt.month

# 계절 정보 추가
def get_season(month):
    if 3 <= month <= 5:
        return '봄'
    elif 6 <= month <= 8:
        return '여름'
    elif 9 <= month <= 11:
        return '가을'
    else:
        return '겨울'

sales_df['계절'] = sales_df['월'].apply(get_season)

# 일별, 월별 구매금액 계산
total_purchase_by_day = sales_df.groupby('거래일자')['구매금액'].sum()
total_purchase_by_month = sales_df.groupby('월')['구매금액'].sum()

print("일별 구매금액:")
print(total_purchase_by_day)

print("\n월별 구매금액:")
print(total_purchase_by_month)

# 주중/주말과 계절에 따른 구매금액 계산
total_purchase_by_weekday = sales_df.groupby('주중여부')['구매금액'].sum()
total_purchase_by_season = sales_df.groupby('계절')['구매금액'].sum()

print("\n주중/주말에 따른 구매금액:")
print(total_purchase_by_weekday)

print("\n계절에 따른 구매금액:")
print(total_purchase_by_season)
```

## 2. 상품별 분석

```python
# 가격대별 상품 분류
def categorize_price(price):
    return f"{price // 10000 * 10000}-{(price // 10000 + 1) * 10000 - 1}원"
sales_df['가격대'] = sales_df['구매금액'].apply(categorize_price)

# 가격대별 상품 분포
price_category_distribution = sales_df['가격대'].value_counts().sort_index()
print("가격대별 상품 분포:")
print(price_category_distribution)
```

```python
# 월별 상품 중분류별 구매금액 추이
sales_df['거래일자'] = pd.to_datetime(sales_df['거래일자'], format='%Y%m%d')

sales_df['월'] = sales_df['거래일자'].dt.month

monthly_purchase_by_category = sales_df.groupby(['월', '상품중분류명'])['구매금액'].sum().unstack()

print("월별 중분류별 구매금액 합계:")
monthly_purchase_by_category

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd

# 한글 폰트 설정 (예: AppleGothic)
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 그래프 설정
plt.figure(figsize=(14, 8))
monthly_purchase_by_category.plot(kind='line', marker='o', figsize=(14, 8))

# 그래프 제목 및 라벨 설정
plt.title('월별 중분류별 구매금액 추이')
plt.xlabel('월')
plt.ylabel('구매금액')
plt.legend(title='중분류명', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()
```

```python
# 주중/ 주말/ 계절별 / 가격대별 상품 구매금액 추이 (집중 판매 상품있는지 확인)
# 거래일자를 날짜형으로 변환
sales_df['거래일자'] = pd.to_datetime(sales_df['거래일자'], format='%Y%m%d')

# 요일 정보 추가 (0: 월요일, 1: 화요일, ..., 6: 일요일)
sales_df['요일'] = sales_df['거래일자'].dt.dayofweek

# 요일을 평일(0~4)과 주말(5,6)로 구분
sales_df['주중여부'] = sales_df['요일'].apply(lambda x: '평일' if x < 5 else '주말')

# 월 정보 추가
sales_df['월'] = sales_df['거래일자'].dt.month

# 계절 정보 추가
def get_season(month):
    if 3 <= month <= 5:
        return '봄'
    elif 6 <= month <= 8:
        return '여름'
    elif 9 <= month <= 11:
        return '가을'
    else:
        return '겨울'

sales_df['계절'] = sales_df['월'].apply(get_season)

# 가격대 정보 추가
def categorize_price(price):
    return f"{price // 10000 * 10000}-{(price // 10000 + 1) * 10000 - 1}원"

sales_df['가격대'] = sales_df['구매금액'].apply(categorize_price)

# 주중/주말별 구매금액 합계 계산
weekday_purchase = sales_df.groupby('주중여부')['구매금액'].sum()
print("주중/주말별 구매금액 합계:")
print(weekday_purchase)

# 계절별 구매금액 합계 계산
season_purchase = sales_df.groupby('계절')['구매금액'].sum()
print("\n계절별 구매금액 합계:")
print(season_purchase)

# 가격대별 구매금액 합계 계산
price_category_purchase = sales_df.groupby('가격대')['구매금액'].sum().sort_index()
print("\n가격대별 구매금액 합계:")
print(price_category_purchase)

# 시각화
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd

# 한글 폰트 설정 (예: AppleGothic)
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지


# 주중/주말별 구매금액 시각화
plt.figure(figsize=(10, 6))
weekday_purchase.plot(kind='bar', color='skyblue')
plt.title('주중/주말별 구매금액')
plt.xlabel('주중/주말')
plt.ylabel('구매금액')
plt.xticks(rotation=0)
plt.show()

# 계절별 구매금액 시각화
plt.figure(figsize=(10, 6))
season_purchase.plot(kind='bar', color='orange')
plt.title('계절별 구매금액')
plt.xlabel('계절')
plt.ylabel('구매금액')
plt.xticks(rotation=0)
plt.show()

# 가격대별 구매금액 시각화
plt.figure(figsize=(14, 8))
price_category_purchase.plot(kind='bar', color='green')
plt.title('가격대별 구매금액')
plt.xlabel('가격대')
plt.ylabel('구매금액')
plt.xticks(rotation=45)
plt.show()
```
